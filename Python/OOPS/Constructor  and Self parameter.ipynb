```python
# self Parameter

class Student:
    def __init__(self,name,Rollno):
        self.name=name
        self.Rollno=Rollno
        self.branch="ECE"
    def show(self):
        print(f"student name:{self.name}")
        print(f"student rollno:{self.Rollno}")
        print(f"Student branch:{self.branch}")
s1=Student("priya",101)
s2=Student("Rama",102)
s1.show()
```
