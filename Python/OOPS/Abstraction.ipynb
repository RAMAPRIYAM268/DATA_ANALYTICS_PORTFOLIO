```python

from abc import ABC,abstractmethod
class Employee(ABC):
    @abstractmethod
    def work(self):
        print("work")
class HR(Employee):
    def work(self):
        print("Managing")
class DataAnalyst(Employee):
    def work(self):
        return "analyse"
        
class Developer(Employee):
    def work(self):
        print("Coding")
hr=HR()
D=Developer()
hr.work()
D.work()
DA1=DataAnalyst()
DA1.work()



```
