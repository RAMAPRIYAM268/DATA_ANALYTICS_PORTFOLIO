```python
# Customer List Management (Lists)
customers = ["John", "Mary", "David"]
print(customers)

#Add a new customer.
new_customer=input("Enter new customer:")
customers.append(new_customer)
print("Adding customer:",customers)

#Remove a customer.
remove_customer=input("Enter customers to remove")
if remove_customer in customers:
    customers.remove(remove_customer)
    print("customer removed",remove_customer)
else:
    print("customer not found")
print(customers)

print("\n ")
#Sort customers alphabetically.
customers.sort()
print("sorted customers",customers)

#Display total number of customers.
print("Total no of customers:",len(customers))

#Display the first and last customer.
print("First customer:",customers[0])
print("Second customer:",customers[-1])
```

```python
#Company Information (Tuples)
company = ("ABC Support Services", "Coimbatore", 2015)

#Display all values.
print("Company Information:",company)

#Access individual elements.
for item in company:
    print(item)
print("\n ")
print("Company Name:",company[0])
print("Location:",company[1])
print("Company Established:",company[2])


#Display company age.
from datetime import datetime
current_year=datetime.now().year
company_age=current_year-company[2]
print("Company Age:",company_age,"Years")
```

```python
# Unique Customer Cities (Sets)
cities = {"Chennai", "Coimbatore", "Madurai", "Chennai"}

#Display all unique cities.
print("Unique cities:",cities)

#Count unique cities.
print("Count Unique cities:",len(cities))

#Add a new city.
new_city=input("Enter new city:")
cities.add(new_city)
print("Adding new city:",cities)

print("\n ")

#Remove a city.
remove_city=input("Enter city to remove:")
if remove_city in cities:
    cities.remove(remove_city)
    print("Removed city:",remove_city)
else:
    print("city does not found")
print(cities)

#Check whether a city exists.
check_city=input("Enter a city to search")
if check_city in cities:
    print("City Exists",check_city)
else:
    print("City does not Exists")
```

```python
#Customer Details (Dictionaries)
customer = {
    "Customer_ID": "C101",
    "Name": "John",
    "City": "Chennai",
    "Satisfaction_Rating": 4.5
}

print("Customer Details:",customer)
print("\n ")

#Display all customer details.
for key,values in customer.items():
    print(key,":",values)

print("\n ")

#Update satisfaction rating.
#Add a new field:
#Membership Type
#Remove a field.

customer["Satisfaction_Rating"]=4.8
customer["Membership Type"]="Gold"
customer.pop("City")
for key,values in customer.items():
    print(key,":",values)


#Display all keys and values separately.

print("\n Keys ")
for key,values in customer.items():
    print(key)

print("\n Values")
for key,values in customer.items():
    print(values)
```

```python
#Business Analysis Tasks
#Generate a simple customer summary report.
print(" Summary Report")
print("\n")

#How many customers are stored in the system?
customer_records=[{"customerID":"C101",
                   "Name":"John",
                   "City":"Chennai",
                   "Rating":4.5
                   },
                   {
                    "customerID":"C102",
                   "Name":"Mary",
                   "City":"Coimbatore",
                   "Rating":4.8
                   },
                   {
                    "customerID":"C103",
                   "Name":"David",
                   "City":"Madurai",
                   "Rating":4.1
                   },
                   {
                    "customerID":"C104",
                   "Name":"Rihana",
                   "City":"Chennai",
                   "Rating":4.4
                   }]

print("Total Customers:",len(customer_records))

#How many unique cities are represented?
unique_cities=set()
for records in customer_records:
    unique_cities.add(records["City"])
print(" Total no of Unique Cities:",len(unique_cities))

print("\n")

#Which customer has the highest satisfaction rating?
highest_customer=customer_records[0]
for records in customer_records:
    if records["Rating"]>highest_customer["Rating"]:
        highest_customer=records
print("Satisfaction Rating:",highest_customer["Name"],"-",highest_customer["Rating"])


print("\n")

#Identify duplicate customer cities and store only unique cities.
print("\n Cities:")
for city in unique_cities:
    print(city)


```

```python
#Bonus Task
#User enters Customer ID.
#Display corresponding customer details.
#If customer not found, display an appropriate message.
Search_customerID=(input("Enter CustomerID"))
found=False
for record in customer_records:
    if record["customerID"]==Search_customerID:
        for key,value in record.items():
            print(key,":",value)
        found=True
        break
if not found:
    print("customer not found")

```
