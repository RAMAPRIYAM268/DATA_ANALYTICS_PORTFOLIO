```python
#Assignment 02 – Employee Attendance Monitoring System
#Attendance Summary
Total_employees=0
Excellent=0
Good=0
Required_improvement=0

#Attendance Classification
attendance_list=[]
for employees in range(1,11):
    attendance=float(input(f"Employee {employees} Attendance:"))

# Using break
    if attendance==-1:
        print("Data entry stop at a time")
        break
#Using continue
    if attendance==0:
        print("Skip the record")
        continue
#Using pass
    if attendance>100:
        pass
#Total Employees Processed
    Total_employees+=1
    attendance_list.append(attendance)

#Number of Excellent Employees
#Number of Good Employees
#Number of Employees Requiring Improvement
    if attendance>=90:
        Category="Excellent"
        Excellent+=1
    elif attendance>75:
        Category="Good"
        Good+=1
    else:
        Category="Improvement Required"
        Required_improvement+=1
    print(f"Employee{employees}:{Category}")

    #Attendance Summary
print("\n Attendance Summary")
print(" ")
print("Total Employees Processed :",Total_employees)
print("Excellent :",Excellent)
print("Good :",Good)
print("Improvement Required:",Required_improvement)


#Bonus Task
if Total_employees>0:
    highest=round(max(attendance_list))
    lowest=round(min(attendance_list))
    average=round(sum(attendance_list)/len(attendance_list))
    Excellent_percentage=round((Excellent/Total_employees)*100)
    Good_percentage=round((Good/Total_employees)*100)
    ImprovementRequired_percentage=round((Required_improvement/Total_employees)*100)


    print("\n Additional Insights")
    print(" ")
    print(f"Highest Attendance Percentage :{highest}%")
    print(f"Lowest Attendance Percentage :{lowest}%")
    print(f"Average Attendance Percentage :{average}%")
#Percentage of Employees in Each Category
    print(f"Excellent Percentage : {Excellent_percentage}%")
    print(f"Good Percentage :{Good_percentage}%")
    print(f"Improvement Required Percentage :{ImprovementRequired_percentage}%")



    


```
