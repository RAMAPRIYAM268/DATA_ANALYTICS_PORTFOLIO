```python
#Assignment 01 – Student Admission Eligibility Checker

#Accept
studentName=input("Enter name")
age=int(input("Enter age"))
degree=input("Enter degree")
percentage=float(input("Enter percentage"))

#Display details
print("\nSTUDENT DETAILS")
print("Name:",studentName)
print("Age:",age)
print("Degree:",degree)
print("Percentage:",percentage)


#Determine Eligibility
if percentage>=60:
    print("The student is Eligible for Admission")
else:
    print("The student is not Eligible for Admission")

#Calculate
Years_30=30-age
percentage_100=100-percentage
print("Years remaining to reach age 30:",Years_30)
print("Percentage required to reach 100:",percentage_100)    
```
