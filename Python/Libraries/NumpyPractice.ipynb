```python
#Array Indexing
a=np.array([["Jan","Feb","March"],
            [300,400,700],
            [4,5,6],
            [200,500,600]])

print(a[0])
```

```python
a=np.array([["Jan","Feb","March"],
            [300,400,700],
            [4,5,6],
            [200,500,600]])
print(a[2][2])
print(a[1,:])
print(a[:,1])
print(a[1:,2])
print(a[2:,1])
print(a[2,1])
```

```python
#Boolean Filtering
s_r=np.array([3.6,4,2,3,1])
print(s_r[s_r>3])
print(s_r)
```

```python
s_r=np.array([3.6,4,2,3,1])
print([s_r>3])
print(s_r[s_r>=2])
```

```python
s_r=np.array([3.6,4,2,3,1,7,9])
f=(s_r>=2) & (s_r<=4)
print(s_r[f])
```

```python
a=np.array([["Jan","Feb","March"],
            [300,400,700],
            [4,5,6],
            [200,500,600]])
print(a[1:,:])
print(np.sum((a[1:,:].astype('int'))))
print(np.mean((a[1:,:].astype('int')))) #Average
print(np.max((a[1:,:].astype('int'))))
print(np.min((a[1:,:].astype('int'))))
print(np.std((a[1:,:].astype('int'))))




```
